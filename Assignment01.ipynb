```python
#Question 2
def Sal(work_hours):
  a = 0
  for x in work_hours:
    if x[-1] == 'F':
      if(a < x[-2]):
        a = x[-2]
        nam = x[-3] 
  return nam,a*20

work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]


name,sal=Sal(work_hours)
print("the female worker that worked the most hours is",name,"and she should be paid",sal)
```
